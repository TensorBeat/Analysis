```
'''
Well, that's useful...
Let's try and do it conventionally
'''

import math

domain = list(range(len(freq)))
f_t = zip(domain, freq)

def fourier(a, b, c):
    return b*math.cos(-c*a), b*math.sin(-c*a)


def wrap_function(yvals, n):
    coords = []
    for i in range(n):
        coords.append(fourier(i, yvals[i], i))
    return coords

circled = wrap_function(freq, 2000000)
plt.axes(projection='polar')
plt.plot(circled)
```
